````python
### Task 3.1 — Cleaning Strategy

# | Column           | Strategy                             | Justification                                                                 |
# |------------------|--------------------------------------|-------------------------------------------------------------------------------|
# | customer_id      | Drop the row if value is missing     | It is a unique value used to identify a customer.                             |
# | customer_name    | Fill with "Unknown"                  | It is not always necessary for analysis compared to their feedback or purchases. |
# | email            | Fill with "Unknown"                  | We need purchase data for analysis, not always the email validity.            |
# | phone            | Fill with "Unknown"                  | We should not remove data just because the phone number is missing/invalid (some customers prefer not to share phone numbers). |
# | registration_date| Fill with mean/median                | To preserve the timeline without losing data.                                 |
# | age              | Fill with median age                 | To avoid inconsistency caused by outliers.                                    |
# | purchase_amount  | Drop missing values                  | It is the main metric for analysis and should not be replaced with random values. |
# | product_category | Fill with mode or "Unknown"          | To maintain consistent category analysis.                                     |
# | payment_method   | Fill with mode or "Unknown"          | It does not significantly affect analysis since it is not the most important metric. |
````

````python
#Task 3.2 - Implement Missing Value Handling
# Create df_clean copy
df_clean = df.copy()

# Drop missing customer_id
df_clean = df_clean[~df_clean['customer_id'].isna()]

# Fill text fields
df_clean['customer_name'] = df_clean['customer_name'].fillna("Unknown")
df_clean['email'] = df_clean['email'].fillna("Unknown")
df_clean['phone'] = df_clean['phone'].fillna("Unknown")

# Registration date fill -> mode
df_clean['registration_date'] = df_clean['registration_date'].fillna(df_clean['registration_date'].mode()[0])

# Age — fixing invalids then median fill
df_clean.loc[(df_clean['age']<0)|(df_clean['age']>120),'age']=np.nan
df_clean['age']=df_clean['age'].fillna(df_clean['age'].median())

# Drop missing purchase amounts as well as negative values
df_clean = df_clean[df_clean['purchase_amount'].notna()]
df_clean = df_clean[df_clean['purchase_amount'] >= 0]
# Mode fill
df_clean['product_category']=df_clean['product_category'].fillna(df_clean['product_category'].mode()[0])
df_clean['payment_method']=df_clean['payment_method'].fillna(df_clean['payment_method'].mode()[0])
````

````python
# Part 5: Data Standardization

# Task 5.1: Standardize Customer Names
# Standardize the customer_name column to Title Case with no extra whitespace.
# Example: "  john doe  " → "John Doe"

#  python
# Your code here

# strip() removes leading/trailing spaces.
# title() converts first letter of every word to uppercase.
df_clean['customer_name'] = (df_clean['customer_name'].str.strip().str.title())

````

````python
# Task 5.2: Standardize Email Addresses
# Standardize the email column to lowercase with no extra whitespace.

df_clean['email'] = (df_clean['email'].str.strip().str.lower())
````

````python
# Task 5.3: Standardize Phone Numbers
# Standardize the phone column to a consistent format: XXX-XXXX
# Example: "(555) 1234", "555.1234", "5551234" → "555-1234"
# Hint: Remove all non-numeric characters first, then format.

# \D removes everything except digits.
# we keep only last 7 digits.
# Then insert "-" after 3 digits.
# astype() -> converts(casts) the datatype

df_clean['phone'] = (df_clean['phone'].astype(str).str.replace(r'\D', '', regex=True).str[-7:].str.replace(r'(\d{3})(\d{4})', r'\1-\2', regex=True))
````

````python
# Task 5.4: Standardize Country Names
# Standardize the country column to uppercase.

# Makes values like india / INDIA / India consistent.
df_clean['country'] = df_clean['country'].str.strip().str.upper()
````

````python
# Task 5.5: Standardize Product Categories
# Standardize the product_category column to Title Case.

df_clean['product_category'] = df_clean['product_category'].str.strip().str.title()
````

````python
# Task 5.6: Standardize Payment Methods
# Standardize the payment_method column to Title Case.

df_clean['payment_method'] = df_clean['payment_method'].str.strip().str.title()
````

````python
# Task 5.7: Verify Standardization
# Display unique values for all categorical columns to verify standardization:

print(df_clean['country'].unique())
print(df_clean['product_category'].unique())
print(df_clean['payment_method'].unique())
````

````python
# Part 6: Data Type Conversion

# Task 6.1: Convert customer_id to Integer
# Convert the customer_id column to integer type.

# astype(int) -> converts the datatype to integer.
df_clean['customer_id'] = df_clean['customer_id'].astype(int)
````

````python
# ============================================================
# Task 6.2: Convert registration_date to Datetime
# ============================================================

# Step 1 — Fill missing dates with the most frequent date (mode)
df_clean['registration_date'] = df_clean['registration_date'].fillna(
    df_clean['registration_date'].mode()[0]
)

# Convert to datetime format
# errors='coerce' turns bad formats into NaT instead of crashing
df_clean['registration_date'] = pd.to_datetime(
    df_clean['registration_date'],
    errors='coerce',
    dayfirst=False     # change to True only if most dates are DD/MM/YYYY
)

# Step 3 — If any NaT appeared again, fill them back with mode
df_clean['registration_date'] = df_clean['registration_date'].fillna(
    df_clean['registration_date'].mode()[0]
)
````

````python
# Task 6.3: Convert Categorical Columns
# Convert country, product_category, and payment_method to categorical type.

# Categorical saves memory and improves performance.

df_clean['country'] = df_clean['country'].astype('category')
df_clean['product_category'] = df_clean['product_category'].astype('category')
df_clean['payment_method'] = df_clean['payment_method'].astype('category')
````

````python
# Task 6.4: Ensure Proper Numeric Types
# Ensure age is integer and purchase_amount is float.

df_clean['age'] = df_clean['age'].astype(int)
df_clean['purchase_amount'] = df_clean['purchase_amount'].astype(float)
````

````python
# Task 6.5: Verify Data Types
# Display the data types of all columns to verify conversions:

df_clean.dtypes
````

````python
# Part 7: Handling Outliers and Invalid Values

# Task 7.1: Fix Invalid Ages
# Identify and fix:
# - Negative ages
# - Ages greater than 120

# Replace invalid ages with the median age.
median_age = df_clean['age'].median()
df_clean.loc[(df_clean['age'] < 1) | (df_clean['age'] > 120), 'age'] = median_age
````

````python
# Task 7.2: Handle Purchase Amount Outliers
# Use the IQR (Interquartile Range) method to:
# 1. Calculate Q1, Q3, and IQR
# 2. Determine lower and upper bounds
# 3. Identify outliers
# 4. Handle outliers by capping them at the bounds
#
# Formula:
# IQR = Q3 - Q1
# Lower Bound = Q1 - 1.5 × IQR
# Upper Bound = Q3 + 1.5 × IQR

Q1 = df_clean['purchase_amount'].quantile(0.25)
Q3 = df_clean['purchase_amount'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Identify outliers
outliers = df_clean[(df_clean['purchase_amount'] < lower) | (df_clean['purchase_amount'] > upper)]

# Cap values
df_clean['purchase_amount'] = df_clean['purchase_amount'].clip(lower, upper)

# **Questions:**
# - How many outliers were identified?
# - What were the outlier values?
# - What is the new range of purchase amounts?

print("Number of outliers identified:", outliers.shape[0])
print("Outlier values:\n", outliers['purchase_amount'])
print("New range:", df_clean['purchase_amount'].min(), "to", df_clean['purchase_amount'].max())

````

````python
# Challenge 1: Email Validation
# Task:
# Create a function to validate email addresses using regular expressions.
# Add a new column `email_valid` that indicates whether each email is valid.
#
# Valid Email Pattern: username@domain.extension

# We will use a regular expression (regex) to check if an email is valid.
# If the email matches the correct pattern, we mark it as True.
# Otherwise, we mark it as False.
#
# re.match() checks whether the text matches the pattern.

import re

# Function definition -> checks if an email is valid
def is_valid_email(email):
    # This function checks whether the email follows the pattern:
    # username@domain.extension
    pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'  # regex rule

    # re.match returns a match object if valid, else None
    return bool(re.match(pattern, str(email)))


# checks for each row and create new column
df_clean['email_valid'] = df_clean['email'].apply(is_valid_email)

# Display results
df_clean[['email', 'email_valid']]
````

````python
# Challenge 3: Advanced Outlier Detection — Z-Score Method

# Formula:
# Z = (value - mean) / standard_deviation
#
# If |Z| > 3 → it is considered an outlier
#

# Z-score => how far a value is from the average.
# A score of 3 => "3 standard deviations away".
# Very rare values are labeled as outliers.
#
# np.mean() calculates average
# np.std() calculates standard deviation

import numpy as np

purchase_mean = np.mean(df_clean['purchase_amount'])
purchase_std = np.std(df_clean['purchase_amount'])

# Calculate Z-score for each purchase amount
df_clean['z_score'] = (df_clean['purchase_amount'] - purchase_mean) / purchase_std

# Flag outliers
z_outliers = df_clean[np.abs(df_clean['z_score']) > 3]

print("Number of Z-score outliers:", z_outliers.shape[0])
print(z_outliers[['customer_name','purchase_amount','z_score']])
````
