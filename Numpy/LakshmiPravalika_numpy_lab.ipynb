````python
## Exercise 6: Real-World Application - Student Grade Analysis (25 points)

# You are given grade data for 5 students across 4 subjects (Math, Science, English, History):

# ```python
# grades = np.array([
#     [85, 90, 78, 88],
#     [92, 88, 95, 90],
#     [76, 82, 80, 75],
#     [88, 85, 90, 92],
#     [95, 92, 88, 94]
# ])
# ```

# **Tasks:**

# 1. **Calculate the average grade for each student** (5 points)
#    - Output should show Student 1: X.XX, Student 2: Y.YY, etc.

# 2. **Calculate the average grade for each subject** (5 points)
#    - Output: Math: X.XX, Science: Y.YY, etc.

# 3. **Find the highest and lowest grades in each subject** (5 points)
#    - Output for each subject: Highest = X, Lowest = Y

# 4. **Identify students with an average grade above 85** (3 points)
#    - Print student number and their average

# 5. **Calculate the overall class average** (2 points)
#    - Average of all grades across all students

# 6. **Normalize all grades to a scale of 0-10** (5 points)
#    - Use min-max normalization formula: (value - min) / (max - min) * 10
#    - Print the normalized grades array

# **Bonus:** Identify the top-performing student (+2 points)

# **Your Code:**
# ```python
import numpy as np

grades = np.array([
    [85, 90, 78, 88],
    [92, 88, 95, 90],
    [76, 82, 80, 75],
    [88, 85, 90, 92],
    [95, 92, 88, 94]
])

# Write your solution here

# 1. Average grade for each student (row-wise mean)
students_avgs = np.mean(grades, axis=1)

print("Average grade per student:")
i = 1
for avg in students_avgs:
    print(f"Student {i}: {avg:.2f}")
    i = i+1

# 2. Average grade for each subject (column-wise mean)
subjects_avgs = np.mean(grades, axis=0)
# for the given subjects Math, Science, English, History
subjects = ["Math", "Science", "English", "History"]
print("\nAverage grade per subject:")
for name, avg in zip(subjects, subjects_avgs): #here zip combines the two things subjects and the students avgs for the subjects
    print(f"{name}: {avg:.2f}")
# 3. Highest and lowest grade in each subject
subject_max = np.max(grades, axis=0) # subject => column wise axis = 0, students => row wise axis =1
subject_min = np.min(grades, axis=0)

print("\nHighest and lowest grades per subject:")
for name, hi, lo in zip(subjects, subject_max, subject_min): # here zip combines the subjects and corresponding max and min values
    print(f"{name}: Highest = {hi}, Lowest = {lo}")

# 4. Students with average above 85
print("\nStudents with average above 85:")
i = 1
for avg in students_avgs:
    if avg > 85:
        print(f"Student {i}: {avg:.2f}")
    i = i+1

# 5. Overall class average (all grades)
class_average = np.mean(grades)
print(f"\nOverall class average: {class_average:.2f}")

# 6. Normalize grades to 0–10 scale using min-max normalization
min = np.min(grades)
max = np.max(grades)

normalized_grades = (grades - min) / (max - min) * 10
print("\nNormalized grades (0–10 scale):")
print(normalized_grades)

# Bonus: Top-performing student (highest average)
top_student = np.argmax(students_avgs) + 1   # used AI to know how to get the index of the a certain value(in this case it's max hence argmax)
                                           # +1 since the indexing starts from 0
print(f"\nTop performing student: Student {top_student}")
````

````python

## Exercise 7: Challenge - Sales Data Analysis (30 points)

# You are analyzing sales data for a company with 4 products over 6 months:

# ```python
# sales_data = np.array([
#     [120, 135, 150, 145, 160, 175],  # Product A
#     [90, 95, 100, 110, 115, 120],    # Product B
#     [150, 145, 155, 160, 170, 180],  # Product C
#     [80, 85, 82, 88, 92, 95]         # Product D
# ])
# ```

# **Tasks:**

# 1. **Calculate total sales for each product** (5 points)
#    - Sum across all months for each product

# 2. **Calculate total sales for each month** (5 points)
#    - Sum across all products for each month (Jan to Jun)

# 3. **Identify the month with the highest total sales** (4 points)
#    - Print the month name and total sales

# 4. **Calculate the growth rate for each product** (6 points)
#    - Growth rate = ((Last month - First month) / First month) Ã— 100
#    - Print as percentage for each product

# 5. **Find products with average monthly sales greater than 120k** (4 points)
#    - Calculate average for each product
#    - Print products meeting the criteria

# 6. **Calculate quarter-wise sales** (6 points)
#    - Q1: January-March
#    - Q2: April-June
#    - Print sales for each product in each quarter

# **Bonus:** Predict next month (July) sales for each product (+5 points)
# - Use a simple method: average of last 2 months + average monthly growth

# **Your Code:**
# ```python
import numpy as np

sales_data = np.array([
    [120, 135, 150, 145, 160, 175],  # Product A
    [90, 95, 100, 110, 115, 120],    # Product B
    [150, 145, 155, 160, 170, 180],  # Product C
    [80, 85, 82, 88, 92, 95]         # Product D
])

products = ["Product A", "Product B", "Product C", "Product D"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

# 1. Total sales per product (row-wise sum) => axis 1
total_per_product = np.sum(sales_data, axis=1)
print("Total sales per product:")
for p, total in zip(products, total_per_product):
    print(f"{p}: {total}")

# 2. Total sales per month (column-wise sum) => axis 0
total_per_month = np.sum(sales_data, axis=0)
print("\nTotal sales per month:")
for m, total in zip(months, total_per_month):
    print(f"{m}: {total}")

# 3. Month with highest total sales
max_month_index = np.argmax(total_per_month) # argmax returns the index of the highest value in matrix
print(f"\nHighest total sales month: {months[max_month_index]} "
      f"({total_per_month[max_month_index]})")

# 4. Growth rate for each product
first_month = sales_data[:, 0] # [:,x] -> all rows but xth column ( here all products(rows) at first month(0th col))
last_month = sales_data[:, -1] #likewise for the last month(Jun)
growth_rate = ((last_month - first_month) / first_month) * 100

print("\nGrowth rate per product (%):")
for p, g in zip(products, growth_rate):
    print(f"{p}: {g:.2f}%")

# 5. Products with average monthly sales > 120
avg_sales = np.mean(sales_data, axis=1)
print("\nProducts with average monthly sales > 120k:")
for p, avg in zip(products, avg_sales):
    if avg > 120:
        print(f"{p}: {avg:.2f}")

# 6. Quarter-wise sales
q1 = np.sum(sales_data[:, 0:3], axis=1)   # Jan–Mar
q2 = np.sum(sales_data[:, 3:6], axis=1)   # Apr–Jun

print("\nQuarter-wise sales:")
i = 0
for p in products:
    print(f"{p} — Q1: {q1[i]}, Q2: {q2[i]}")
    i=i+1

# BONUS — Predict July sales
# Method: (avg of last 2 months) + (average monthly growth)

last_two_avg = np.mean(sales_data[:, -2:], axis=1) # : => indicate all products , -2: => indicate from the last 2 months
monthly_growth = (sales_data[:, -1] - sales_data[:, -2])  # growth from May->Jun

july_prediction = last_two_avg + monthly_growth

print("\nPredicted July sales (simple model):")
for p, pred in zip(products, july_prediction):
    print(f"{p}: {pred:.2f}")
````
