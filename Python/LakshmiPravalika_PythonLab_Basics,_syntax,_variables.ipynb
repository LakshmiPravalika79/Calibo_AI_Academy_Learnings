````python
# Python Essentials Lab - Student Exercises

## Lab Instructions

"""Complete each exercise by writing Python code that meets the requirements. Test your code thoroughly and ensure it produces the expected output format.

**Submission Guidelines**:
- Write clean, readable code with proper indentation
- Use meaningful variable names
- Add comments to explain your logic
- Test your code with different inputs

---"""

## Exercise 1: Variable Declaration and Basic Operations

"""**Objective**: Practice creating variables and performing basic operations."""

### Requirements

"""Create a program that stores and displays student information:

1. Create variables for the following information:
   - Student name: "Sarah Johnson"
   - Age: 20
   - GPA: 3.75
   - Is enrolled: True

2. Calculate the student's age after 5 years

3. Display all information in a formatted output

### Expected Output Format
```
Student Profile
========================================
Name: Sarah Johnson
Current Age: 20
Age in 5 years: 25
GPA: 3.75
Enrolled: True
```

### Write Your Code Here
```python
"""
# Your code here

# Created variables to store student information
student_name = "Sarah Johnson"
age = 20
gpa = 3.75
is_enrolled = True

# To calculate the student's age after 5 years
age_after_5_years = age + 5

# Displaying the student profile in a formatted way
print("Student Profile")
print("========================================")
print(f"Name: {student_name}")
print(f"Current Age: {age}")
print(f"Age in 5 years: {age_after_5_years}")
print(f"GPA: {gpa}")
print(f"Enrolled: {is_enrolled}")

## Exercise 2: Type Conversion and User Input

"""**Objective**: Practice getting user input and converting between data types.

### Requirements
Create a program that:

1. Asks the user to enter their name (string)
2. Asks for their birth year (convert to integer)
3. Asks for their height in meters (convert to float)
4. Calculates their age (assume current year is 2025)
5. Displays all information including the data type of each variable

### Expected Output Format
```
Enter your name: Alex
Enter your birth year: 2000
Enter your height in meters: 1.75

========================================
User Information
========================================
Name: Alex
Birth Year: 2000
Age: 25 years old
Height: 1.75 meters
Name type: <class 'str'>
Age type: <class 'int'>
Height type: <class 'float'>
```

### Write Your Code Here
```python
"""
# Your code here

# Ask the user for their name (by default string)
name = input("Enter your name: ")

# Ask for birth year and conversion of input to an integer
birth_year = int(input("Enter your birth year: "))

# Ask for height in meters and conversion of input to a float
height = float(input("Enter your height in meters: "))

# Assume current year is 2025
current_year = 2025

# Calculate age
age = current_year - birth_year

# Display the information
print("\n========================================")
print("User Information")
print("========================================")
print(f"Name: {name}")
print(f"Birth Year: {birth_year}")
print(f"Age: {age} years old")
print(f"Height: {height} meters")

# Display the data types of each variable
print(f"Name type: {type(name)}")
print(f"Age type: {type(age)}")
print(f"Height type: {type(height)}")

## Exercise 3: Arithmetic Operations and Calculations

"""**Objective**: Practice using arithmetic operators.

### Requirements
Create a simple calculator program that:

1. Takes two numbers as input from the user
2. Performs the following operations:
   - Addition
   - Subtraction
   - Multiplication
   - Division
   - Floor division
   - Modulus
   - Exponentiation
3. Displays all results with proper labels

### Expected Output Format
```
Enter first number: 15
Enter second number: 4

========================================
Calculator Results
========================================
15.0 + 4.0 = 19.0
15.0 - 4.0 = 11.0
15.0 Ã— 4.0 = 60.0
15.0 Ã· 4.0 = 3.75
15.0 Ã·Ã· 4.0 = 3.0
15.0 % 4.0 = 3.0
15.0 ^ 4.0 = 50625.0
```"""

### Write Your Code Here
"""```python"""
# Your code here

# Get two numbers from the user and convert them to float(for the expected results as sample output)
first_number = float(input("Enter first number: "))
second_number = float(input("Enter second number: "))

# Perform arithmetic operations
addition = first_number + second_number
subtraction = first_number - second_number
multiplication = first_number * second_number


# Exponentiation (** is the power operator in Python)
exponent = first_number ** second_number

# Display results
print("\n========================================")
print("Calculator Results")
print("========================================")
print(f"{first_number} + {second_number} = {addition}")
print(f"{first_number} - {second_number} = {subtraction}")
print(f"{first_number} × {second_number} = {multiplication}")

# Handle division safely in case of divided by zero
if second_number != 0:
    division = first_number / second_number
    floor_division = first_number // second_number
    modulus = first_number % second_number
  # :.2f results in displaying only upto 2 decimal places of a float value
    print(f"{first_number:.2f} ÷ {second_number:.2f} = {division:.2f}")
    print(f"{first_number:.2f} // {second_number:.2f} = {floor_division:.2f}") # // for floor value is the integer value when divided
    print(f"{first_number:.2f} % {second_number:.2f} = {modulus:.2f}")
else:
    print("Division, floor division, and modulus are undefined (division by zero)")
print(f"{first_number} ^ {second_number} = {exponent}")  #  ^ for power of()



## Exercise 4: String Manipulation

"""**Objective**: Practice string operations and methods.

### Requirements
Create a program that:

1. Takes a sentence as input from the user
2. Displays the sentence in:
   - Uppercase
   - Lowercase
   - Title case
3. Counts and displays:
   - Number of characters
   - Number of words
4. Asks for a word to replace and the new word
5. Displays the modified sentence

### Expected Output Format
```
Enter a sentence: Python is an amazing programming language
Enter a word to replace: amazing
Enter the new word: powerful

========================================
String Analysis
========================================
Original: Python is an amazing programming language
Uppercase: PYTHON IS AN AMAZING PROGRAMMING LANGUAGE
Lowercase: python is an amazing programming language
Title Case: Python Is An Amazing Programming Language
Character count: 41
Word count: 6
Modified: Python is an powerful programming language
```
"""
### Write Your Code Here
"""```python"""
# Your code here

# 1. Take a sentence as input
sentence = input("Enter a sentence: ")


# 4. Ask for word replacement inputs
word_to_replace = input("Enter a word to replace: ")
new_word = input("Enter the new word: ")


# Output section
print("\n========================================")
print("String Analysis")
print("========================================")
print(f"Original: {sentence}")
print(f"Uppercase: {sentence.upper()}") # 2. input sentence in uppercase
print(f"Lowercase: {sentence.lower()}") # 2. input sentence in lower case
print(f"Title Case: {sentence.title()}") # 2. input sentence in title case (i.e, First letter capital in every word)
print(f"Character count: {len(sentence)}") # character count of input sentence
print(f"Word count: {len(sentence.split())}") # word count in the input sentence
# 5. Replace the word
sentence = sentence.replace(word_to_replace, new_word)
print(f"Modified: {sentence}")


## Exercise 5: Temperature Converter

"""**Objective**: Combine variables, input, calculations, and output.

### Requirements
Create a temperature converter that:

1. Asks for temperature in Celsius
2. Converts to Fahrenheit using: (Celsius Ã— 9/5) + 32
3. Converts to Kelvin using: Celsius + 273.15
4. Displays all three temperatures with 2 decimal places

### Expected Output Format
```
Enter temperature in Celsius: 25

========================================
Temperature Conversion
========================================
Celsius:    25.00Â°C
Fahrenheit: 77.00Â°F
Kelvin:     298.15K
```
"""
### Write Your Code Here
"""```python"""
# Your code here

# Ask to enter a temperature in Celsius
celsius = float(input("Enter temperature in Celsius: "))

# Convert Celsius to Fahrenheit using the formula
fahrenheit = (celsius * 9/5) + 32

# Convert Celsius to Kelvin
kelvin = celsius + 273.15

print("\n" + "=" * 40)
print("Temperature Conversion")
print("=" * 40)

# Display all temperatures to 2 decimal places
print(f"Celsius:    {celsius:.2f}°C")
print(f"Fahrenheit: {fahrenheit:.2f}°F")
print(f"Kelvin:     {kelvin:.2f}K")


## Exercise 6: Variable Swapping and Multiple Assignment

"""**Objective**: Practice advanced variable assignment techniques.

### Requirements
Write a program that demonstrates:

1. Multiple assignment with the same value to three variables (a, b, c = 10)
2. Multiple assignment with different values (x=5, y=10, z=15)
3. Swapping two variables without using a temporary variable
4. Display the values before and after each operation

### Expected Output Format
```
Multiple assignment (same value):
a = 10, b = 10, c = 10

Multiple assignment (different values):
x = 5, y = 10, z = 15

Before swap:
x = 5, y = 10
After swap:
x = 10, y = 5
```"""

### Write Your Code Here
"""```python"""
# Your code here

print("\n" + "=" * 40)
print("Variable Swapping and Multiple Assignment")
print("=" * 40)
# 1. Multiple assignment with the same value
a = b = c = 10
print("Multiple assignment (same value):")
print(f"a = {a}, b = {b}, c = {c}\n")

# 2. Multiple assignment with different values
x, y, z = 5, 10, 15
print("Multiple assignment (different values):")
print(f"x = {x}, y = {y}, z = {z}\n")

# 3. Swapping variables without a temporary variable
print("Before swap:")
print(f"x = {x}, y = {y}")

# Python swap trick (tuple unpacking)
x, y = y, x

print("After swap:")
print(f"x = {x}, y = {y}")

## Exercise 7: Shopping Cart Calculator

"""**Objective**: Create a practical application using all concepts learned.

### Requirements
Create a shopping cart program that:

1. Takes input for 3 products (name and price for each)
2. Calculates the subtotal
3. If subtotal is greater than $100, applies a 10% discount
4. Adds 8% tax on the discounted amount
5. Displays an itemized receipt showing:
   - Each product with price
   - Subtotal
   - Discount (if applicable)
   - Tax amount
   - Total amount

**Hint**: Use constants for tax rate (0.08), discount threshold (100), and discount rate (0.10)

### Expected Output Format
```
Enter product information:
Product 1 name: Laptop
Product 1 price: $899.99
Product 2 name: Mouse
Product 2 price: $25.50
Product 3 name: Keyboard
Product 3 price: $75.00

========================================
SHOPPING CART RECEIPT
========================================
Laptop                     $  899.99
Mouse                      $   25.50
Keyboard                   $   75.00
----------------------------------------
Subtotal:                  $ 1000.49
Discount (10%):            $ -100.05
After Discount:            $  900.44
Tax (8%):                  $   72.04
========================================
TOTAL:                     $  972.48
========================================
```
"""
### Write Your Code Here
"""```python"""
# Your code here

# Constants [are represented in uppercase only]
TAX_RATE = 0.08               # percentage taken in decimal for easy calculation
DISCOUNT_THRESHOLD = 100      # discount only if subtotal is greater than this
DISCOUNT_RATE = 0.10          # discount taken in decimal for easy calculation

print("Enter product information:")

# Input of 3 products
name1 = input("Product 1 name: ")
price1 = float(input("Product 1 price: $"))

name2 = input("Product 2 name: ")
price2 = float(input("Product 2 price: $"))

name3 = input("Product 3 name: ")
price3 = float(input("Product 3 price: $"))

# calculating the total price of the purchased products
subtotal = price1 + price2 + price3

# applying discount only if subtotal is greater than threshold(discount percentage)
if subtotal > DISCOUNT_THRESHOLD:
    discount = subtotal * DISCOUNT_RATE
else:
    discount = 0

after_discount = subtotal - discount

# Tax after discount
tax = after_discount * TAX_RATE

# Final total
total = after_discount + tax

print("\n" + "=" * 40)
print("SHOPPING CART RECEIPT")
print("=" * 40)

print(f"{name1:<25} $ {price1:8.2f}") # :< is used to left align and 25 implies the word takes 25 spaces for aligning receipt properly
print(f"{name2:<25} $ {price2:8.2f}")
print(f"{name3:<25} $ {price3:8.2f}")

print("-" * 40)

print(f"Subtotal:{'':17} $ {subtotal:8.2f}") # :17 - just padding 17 empty spaces before the total, for reciept to be similar to desired output

if discount > 0:
    discount = -discount
    print(f"Discount (10%):{'':11} $ {discount:8.2f}")
    print(f"After Discount:{'':11} $ {after_discount:8.2f}")

print(f"Tax (8%):{'':17} $ {tax:8.2f}")

print("=" * 40)
print(f"TOTAL:{'':20} $ {total:8.2f}")
print("=" * 40)



## Self-Assessment Checklist

"""Before submitting your lab, verify:

- [ ] All variables are named using proper conventions (snake_case)
- [ ] Data types are used correctly (int, float, str, bool)
- [ ] Type conversions are applied where needed
- [ ] All calculations produce correct results
- [ ] Output is formatted clearly and matches expected format
- [ ] Code is properly indented
- [ ] Comments explain your logic
- [ ] Code has been tested with different inputs

---

## Bonus Challenges

If you complete all exercises, try these additional challenges:

1. **Input Validation**: Add checks to ensure user input is valid (e.g., numeric values are actually numbers)

2. **Enhanced Calculator**: Add a menu to let users choose which operation to perform

3. **Receipt Enhancement**: Add the ability to calculate change if the customer pays with cash

4. **String Analysis**: Count specific characters or check if a word exists in the sentence

5. **Unit Converter**: Expand the temperature converter to include more units (Fahrenheit to Celsius, etc.)

---

## Need Help?

If you're stuck:
1. Review the lecture materials on variables, data types, and operators
2. Check the Python documentation for string and numeric methods
3. Test your code with simple examples first
4. Use `print()` statements to debug your code
5. Ask your instructor for guidance"""

print("Bonus Challenges executed seperately in the below to make it easy to differentiate between previous challenges anf the enhanced onces")
````

````python
# Exercise 3 + Bonus: Calculator with Menu & Validation

# Function to get a number
# Keeps asking until a valid numeric input is entered
def get_float(prompt):
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Please enter a valid number!") # Error message for invalid input

print("Calculator Menu")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
print("5. Floor Division")
print("6. Modulus")
print("7. Exponent")

choice = input("Choose an operation (1–7): ")

first_number = get_float("Enter first number: ")
second_number = get_float("Enter second number: ")

print("\n========================================")
print("Calculator Results")
print("========================================")

if choice == "1":
    print(f"{first_number} + {second_number} = {first_number + second_number}")

elif choice == "2":
    print(f"{first_number} - {second_number} = {first_number - second_number}")

elif choice == "3":
    print(f"{first_number} × {second_number} = {first_number * second_number}")

elif choice == "4":
  # Avoid division by zero
    if second_number != 0:
        print(f"{first_number:.2f} ÷ {second_number:.2f} = {(first_number / second_number):.2f}")
    else:
        print("Division is undefined (division by zero)")

elif choice == "5":
  # Avoid floor division by zero
    if second_number != 0:
        print(f"{first_number:.2f} // {second_number:.2f} = {(first_number // second_number):.2f}")
    else:
        print("Floor division is undefined (division by zero)")

elif choice == "6":
  # Avoid modulo by zero
    if second_number != 0:
        print(f"{first_number:.2f} % {second_number:.2f} = {(first_number % second_number):.2f}")
    else:
        print("Modulus is undefined (division by zero)")

elif choice == "7":
    print(f"{first_number} ** {second_number} = {first_number ** second_number}")

else:
    print("Invalid choice")
````

````python
# Exercise 7 + Bonus Challenge: Shopping Cart with Change & Validation

# Constants [are represented in uppercase only]
TAX_RATE = 0.08               # percentage taken in decimal for easy calculation
DISCOUNT_THRESHOLD = 100      # discount only if subtotal is greater than this
DISCOUNT_RATE = 0.10          # discount taken in decimal for easy calculation

# Keeps asking until a valid price input is entered
def get_price(prompt):
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Enter a valid price!") # Error shown for invalid input

print("Enter product information:")
# Input of 3 products
name1 = input("Product 1 name: ")
price1 = get_price("Product 1 price: $")

name2 = input("Product 2 name: ")
price2 = get_price("Product 2 price: $")

name3 = input("Product 3 name: ")
price3 = get_price("Product 3 price: $")

subtotal = price1 + price2 + price3
# applying discount only if subtotal is greater than threshold(discount percentage)
if subtotal > DISCOUNT_THRESHOLD:
    discount = subtotal * DISCOUNT_RATE
else:
    discount = 0

after_discount = subtotal - discount

# Tax after discount
tax = after_discount * TAX_RATE

# Final total
total = after_discount + tax

cash = get_price("\nAmount paid by customer: $")

print("\n" + "=" * 40)
print("SHOPPING CART RECEIPT")
print("=" * 40)

print(f"{name1:<25} $ {price1:8.2f}") # :< is used to left align and 25 implies the word takes 25 spaces for aligning receipt properly
print(f"{name2:<25} $ {price2:8.2f}")
print(f"{name3:<25} $ {price3:8.2f}")
print("-" * 40)

print(f"Subtotal:{'':17} $ {subtotal:8.2f}") # :17 - just padding 17 empty spaces before the total, for reciept to be similar to desired output
print(f"Discount (10%):{'':11} $ {-discount:8.2f}")
print(f"After Discount:{'':11} $ {after_discount:8.2f}")
print(f"Tax (8%):{'':17} $ {tax:8.2f}")
print("=" * 40)
print(f"TOTAL:{'':20} $ {total:8.2f}")
print("=" * 40)

if cash < total:
    print(f"Not enough cash paid!\nBalance Amount to pay = $ {total-cash}") # if cash paid is not enough to pay total
else:
    change = cash - total
    print(f"Change Due:{'':15} $ {change:8.2f}") # if paid extra than the total
````
