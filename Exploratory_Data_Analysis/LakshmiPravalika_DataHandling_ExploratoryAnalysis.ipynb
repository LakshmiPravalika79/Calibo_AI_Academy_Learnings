````python
## Part 2: Data Cleaning (20 points)

### Task 2.1: Convert Data Types
# Convert the `date` column to datetime format.

# ```python
# Your code here
# ```
df["date"] = pd.to_datetime(df["date"]) #to_datetime() is used to convert a specific column format into a datetime format
df.dtypes #for debugging and verifying the conversion was successful
````

````python
### Task 2.2: Identify Missing Values
# - Check for missing values in all columns
missing_values = df.isnull().sum() # returns the count of null values(missing) for each column
# - Calculate the percentage of missing values for each column
missing_percentage = (missing_values / df.shape[0]) * 100 # count of null values for each column/total no fo entries(rows) * 100 gives percentage
# - Display your findings`
#converted the results into table format using pandas dataframe for display
missing_df = pd.DataFrame({'No of Missing Values': missing_values, 'Percentage %': missing_percentage})
print(f"Missing Values: \n{missing_df}")
````

````python
### Task 2.3: Handle Missing Values
# Based on your findings:
# - Fill missing values in the `age` column with an appropriate measure of central tendency
df["age"].fillna(df["age"].median()) #fills the missing values of age with the median value directly to the existing dataframe
# - Fill missing values in the `gender` column with an appropriate value
df["gender"].fillna(df["gender"].mode()[0])
# - Verify that all missing values have been handled
df.isnull().sum()
# **Question:** Why did you choose that particular method for handling missing values?
# Median works well in dealing with outliers, making it ideal for age.
# Mode ca be use for categorical variables like gender.
````

````python
### Task 2.4: Check for Duplicates
# Check if there are any duplicate rows in the dataset and report your findings.
duplicates = df.duplicated().sum()
print("Number of duplicate rows:", duplicates)
#There are no duplicate rows in the dataset
````

````python
## Part 3: Summary Statistics (20 points)

### Task 3.1: Descriptive Statistics
# Generate and display descriptive statistics for all numerical columns.
df.describe()
````

````python
### Task 3.2: Calculate Additional Statistics
# For the numerical columns (age, quantity, price_per_unit, total_amount), calculate:
numerical_cols = ["age", "quantity", "price_per_unit", "total_amount"]

# .skew() -> Skewness
# .std() -> Standard deviation
stats_df = pd.DataFrame({
    "Skewness": df[numerical_cols].skew(),
    "Standard Deviation": df[numerical_cols].std()
})

stats_df

# **Question:** What does the skewness tell you about these distributions?
# there are right and left tailed distributions in which positive skew implies a right tailed distribution
#  total amount is right skewed due to high value purchases
````

````python
### Task 3.3: Categorical Variable Analysis
# For categorical variables (product_category and gender):

# - Display value counts
for col in ["product_category", "gender"]:
    print(f"\n{col} Value Counts:\n")
    print(df[col].value_counts())

# - Calculate percentage distribution
    print(f"\nPercentage Distribution of {col} :")
    print(df[col].value_counts(normalize=True) * 100) #normalize=True returns a value range(0-1) and *100 makes it a percentage value
````

````python
## Part 4: Outlier Detection (20 points)

### Task 4.1: Implement IQR Method
# Write a function to detect outliers using the IQR (Interquartile Range) method:
# - Calculate Q1 (25th percentile) and Q3 (75th percentile)
# - Calculate IQR = Q3 - Q1
# - Define outliers as values < Q1 - 1.5Ã—IQR or > Q3 + 1.5Ã—IQR

# def detect_outliers_iqr(data, column):
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return data[(data[column] < lower) | (data[column] > upper)]
````

````python
### Task 4.2: Detect Outliers
# Apply your function to detect outliers in:
# - quantity
# - price_per_unit
# - total_amount

# Report the number of outliers in each column.
for col in ["quantity", "price_per_unit", "total_amount"]:
    outliers = detect_outliers_iqr(df, col)
    print(f"{col} outliers:", len(outliers))
````

````python
### Task 4.3: Visualize Outliers
# Create box plots for the three numerical variables to visualize outliers.
plt.figure(figsize=(12,4)) #creates a graph size with 10 inches width and 5 inches hieght for better plotting
for i, col in enumerate(["quantity", "price_per_unit", "total_amount"], 1):
    plt.subplot(1,3,i) # to create subplot of 1 row and three columns of grid
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")

plt.tight_layout()
plt.show()

# **Question:** Should you remove these outliers? Why or why not?
# No, the outliers in the total amount should not be removed as they can be most important for business analytics
````

````python
## Part 5: Distribution Analysis (25 points)

### Task 5.1: Create Histograms
# Create histograms for:
# - age
# - quantity
# - price_per_unit
# - total_amount
# Arrange them in a 2Ã—2 grid.
df[numerical_cols].hist(bins=20, figsize=(10,8)) # all these age,quantity,price_per_unit, and total_amount are already declared as numerical_columns
plt.tight_layout()
plt.show()
````

````python
### Task 5.2: Interpret Distributions
# For each variable, describe:
# - Is it normally distributed?
# - Is it skewed (left or right)?
# - Are there any unusual patterns?

# **Your interpretations:**
# ```
# Age: slight right skewed as there are only fewer older customers

# Quantity: right skewed as bulk purchses are less repeated

# Price per Unit: right skewed as low price products are purchased more

# Total Amount:  right skewed as high values significantly extend the right tail of distribution
# ```
````

````python
### Task 5.3: Test for Normality
# Use the Shapiro-Wilk test to formally test if the distributions are normal. (p-value > 0.05 suggests normal distribution)
for col in numerical_cols:
    stat, p = stats.shapiro(df[col])
    print(f"{col} -> p-value: {p}")
````

````python
## Part 6: Correlation Analysis (20 points)

### Task 6.1: Calculate Correlation Matrix
# Calculate the correlation matrix for numerical variables (age, quantity, price_per_unit, total_amount).
corr_matrix = df[numerical_cols].corr()
corr_matrix
````

````python
### Task 6.2: Visualize Correlations
# Create a heatmap to visualize the correlation matrix with:
# - Annotations showing correlation values
# - Appropriate color scheme
# - Proper title
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm") #annot=True for displayinfg the corr values inside each cell, cmap="coolwarm" to use color gradients to show +ve and -ve correlations
plt.title("Correlation Heatmap")
plt.show()
````

````python
### Task 6.3: Identify Strong Correlations
# Identify and list all pairs of variables with correlation coefficient |r| > 0.5
strong_corr = corr_matrix[(corr_matrix > 0.5) | (corr_matrix < -0.5)]
strong_corr

# **Question:** What does the correlation between total_amount and quantity tell you? Is this expected?
# total amount correlates with quantity strongly and is explected because total amount is calculated based on the quantity and the price factors
````

````python
## Part 7: Categorical Analysis (25 points)

### Task 7.1: Product Category Visualization
# Create a bar chart showing the distribution of transactions across product categories.

df["product_category"].value_counts().plot(kind="bar") #plots each prosuct category with the it's respective counts
plt.title("Transactions by Product Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
````

````python
### Task 7.2: Gender Distribution
# Create a pie chart showing the gender distribution in the dataset.
df["gender"].value_counts().plot(kind="pie", autopct="%1.1f%%") # displays the gender percentages in dataset as a pie chart
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
````

````python
## Part 9: Time Series Analysis (20 points)

### Task 9.1: Extract Date Components
# Extract the following from the date column:
# - year
# - month
# - day
# - day_of_week (name of the day)

df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day
df["day_of_week"] = df["date"].dt.day_name()
````

````python
### Task 9.2: Daily Sales Trend
# Calculate total sales for each day and create a line plot showing the daily sales trend.
daily_sales = df.groupby("date")["total_amount"].sum()

daily_sales.plot(figsize=(10,5))
plt.title("Daily Sales Trend")
plt.ylabel("Total Sales")
plt.show()
````

````python
### Task 9.3: Sales by Day of Week
# Calculate and visualize total sales by day of week using a bar chart.

df.groupby("day_of_week")["total_amount"].sum().plot(kind="bar")
plt.title("Sales by Day of Week")
plt.show()

# **Question:** Are there any patterns in sales across different days of the week?
# Friday and Sunday have the highest sales and on saturday purchases are relatively lesser while other days of week are moderate
````
